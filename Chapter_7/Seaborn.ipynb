# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names() #dataset có sẵn trong seaborn
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#Draw Scatter Plot
sns.scatterplot(data=tips, x='total_bill', y='tip') 
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex') 
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker', size='size') 
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker', size='size', palette='Set2') 
plt.show()
```

```python
#Histogram
np.random.seed(0)
data = np.random.normal(size=100)
sns.histplot(data)
plt.show()
```

```python
sns.histplot(data, bins=5) #bins là số lượng cột
sns.histplot(data, bins=5, kde=True) #kde là đường cong
```

```python
sns.histplot(tips, x = 'size', bins=6, multiple='stack', hue='sex', kde=True)
```
