# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
```

```python
x_point = np.array([1,2,3,4,6])
y_point = np.array([1,2,3,5,6])
```

```python
#Draw Line
plt.plot(x_point, y_point)
plt.xlabel("X_Label")
plt.ylabel("Y_Label")
plt.title("Matplotlib")
plt.show()
```

```python
#Draw Scatter
plt.scatter(x_point, y_point)
plt.xlabel("X_Label")
plt.ylabel("Y_Label")
plt.title("Matplotlib")
plt.show()
```

```python
#Example
year = ([1, 2, 3, 4, 5])
salary = [1000, 2000, 3000, 4000, 5000]
plt.plot(year, salary)
plt.scatter(year, salary)
plt.xlabel("Year")
plt.ylabel("Salary")
plt.grid(True)
plt.title("Salary")
plt.legend(["Salary (USD)", "Year"])
plt.show()

```
